# NUMBER 1

```python
from sympy import *
x, y, z = symbols('x y z')
init_printing(use_latex = True)
```

```python
%matplotlib inline
expr1 = x+1
expr2 = x**2
h = plot( expr1, expr2, xlabel="$x$", ylabel = "$f(x)$" )
```

```python
array = solve(x**2 - x - 1)
array
```

```python
integrate((expr1-expr2), (x, array[1], array[0])).evalf()
```

# NUMBER 2

$$\lim_{y\to7} \frac{y^2-4y-21}{3y^2-17y-28}$$

```python
expr3 = (y**2-4*y-21)/(3*y**2-17*y-28)
limit( expr3, y, 7 )
```

# NUMBER 3

$$\int_{x}^{9} \frac{x+1}{x+2\sqrt{x}-3}$$

```python
expr4 = (x+1)/(x+2*sqrt(x)-3)
integrate (expr4, (x, 7, 9)).evalf()
```
