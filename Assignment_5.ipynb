## Sine Function

$sin(x) = \sum\limits_{k=0}^\infty \frac{(-1)^k}{(2k+1)!}x^{2k+1}$

```python
import numpy as np
import matplotlib.pyplot as plt

size = 1000

def fact(n):
    x = 1
    for i in range(1, n+1):
        x *= i
    return x

def sin(x, n):
    total = 0
    for i in range (n):
        total += ((-1)**i /fact(2*i+1))* x ** (2*i+1)
    return total

x = np.linspace(-10*np.pi, 10*np.pi, size)
y5 = sin(x,5)
y9 = sin(x,9)
y15 = sin(x,15)


plt.figure(1)
plt.clf()
ax = plt.subplot()
ax.plot(x, y5, 'r')
ax.plot(x, y9, 'g')
ax.plot(x, y15, 'b')
ax.plot(x, np.sin(x),'black')
plt.ylim([-1.1, 1.1])
plt.xlim([-15, 15])
plt.show()
```
