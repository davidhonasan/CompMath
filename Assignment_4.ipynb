## Compute following area.

```python
from numpy import random, sqrt, sin, deg2rad, pi
import matplotlib.pyplot as plt

n = 100000
R = 1
x = random.uniform(-1, 1, n)
y = random.uniform(-1, 1, n)

r = sqrt(x**2 + y**2)
index = r < R
x = x[index]
y = y[index]

ylim = sin((30/180)*pi)
indexbelow = y < ylim

circle = pi * (R ** 2)
triangle = (R ** 2)/2 * sin(deg2rad(120))
sector = 120/360 * circle
theory = circle - (sector - triangle)

area = len(y[y < ylim])/n * 4

print('Area = ', area)
print('Theoritical = ', theory)

plt.figure(1)
plt.clf()
ax = plt.subplot()
ax.plot(x, y, '.r')
ax.plot(x[indexbelow], y[indexbelow], '.g')
```
